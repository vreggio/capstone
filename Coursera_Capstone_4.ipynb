```python
import pandas as pd
import numpy as np
```

Read all the HTML page

```python
wiki_postal_canada = pd.read_html('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M')
```

Select the table needed

```python
table = wiki_postal_canada[0]
table
```

Clean the table dropping the 'Not assigned' items

```python
table2 = table[table.Borough != 'Not assigned']
table2
```

Reset Index to start at "0"

```python
table3 = table2.reset_index(drop=True)
table3
```

```python
table3.shape
```

read the CSV with Coordinates

```python
geocord = pd.read_csv('Geospatial_Coordinates.csv', sep=",")
geocord
```

convert the series to DataFrame

```python
geocord2 = pd.DataFrame(data=geocord)
geocord2
```

merge both DataFrames selecting the Postal Code as the junction

```python
answer2 = pd.merge(table3, geocord2, on=['Postal Code'])
answer2
```

let´s see how many boroughs and neighborhoods are in the dataframe

```python
print('The dataframe has {} boroughs and {} neighborhoods.'.format(
        len(answer2['Borough'].unique()),
        answer2.shape[0]
    )
)
```

get the latitude e longitude from Toronto

```python
from geopy.geocoders import Nominatim # convert an address into latitude and longitude values

address = 'Toronto, ON'

geolocator = Nominatim(user_agent="toronto_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geographical coordinate of Toronto are {}, {}.'.format(latitude, longitude))
```

load the folium library

```python
import folium
```

create map of Toronto using latitude and longitude values

```python
map_toronto = folium.Map(location=[latitude, longitude], zoom_start=10)
map_toronto
```

add markers to map of toronto

```python
for lat, lng, borough, neighborhood in zip(answer2['Latitude'], answer2['Longitude'], answer2['Borough'],answer2['Neighbourhood']):
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_toronto)  
    
map_toronto
```

filter dataframe with the word "Toronto" in borough

```python
toronto_data = answer2[answer2['Borough'].str.contains('Toronto')].reset_index(drop=True)
toronto_data
```

create a map with only this clusters

```python
map_toronto2 = folium.Map(location=[latitude, longitude], zoom_start=10)
for lat, lng, borough, neighborhood in zip(toronto_data['Latitude'], toronto_data['Longitude'], toronto_data['Borough'],toronto_data['Neighbourhood']):
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_toronto2)  
    
map_toronto2
```
